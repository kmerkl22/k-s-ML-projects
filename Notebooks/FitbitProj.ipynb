```python
# extra code â€“ computes the data for Figure 2â€“10

def calorie_cat_proportions(data3):
    return data3["Calories_cat"].value_counts() / len(data3)

train_set, test_set = train_test_split(data3, test_size=0.2, random_state=42)

strat_train_set, strat_test_set = train_test_split(data3, test_size=0.2, stratify=data3["Calories_cat"])

compare_props = pd.DataFrame({
    "Overall %": calorie_cat_proportions(data3),
    "Stratified %": calorie_cat_proportions(strat_test_set),
    "Random %": calorie_cat_proportions(test_set),
}).sort_index()
compare_props.index.name = "Calorie Category"
compare_props["Strat. Error %"] = (compare_props["Stratified %"] /
                                   compare_props["Overall %"] - 1)
compare_props["Rand. Error %"] = (compare_props["Random %"] /
                                  compare_props["Overall %"] - 1)
(compare_props * 100).round(2)
```
